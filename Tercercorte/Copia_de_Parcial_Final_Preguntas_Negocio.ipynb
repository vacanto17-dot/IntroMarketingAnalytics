2. Outliers

```python
import pandas as pd

def detectar_outliers(df, columna):
    if columna not in df.columns or not pd.api.types.is_numeric_dtype(df[columna]):
        print(f"Advertencia: La columna '{columna}' no existe o no es num√©rica. Saltando detecci√≥n de outliers.")
        return pd.DataFrame(), None # Return empty DataFrame and None for limit

    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_superior = Q3 + 1.5 * IQR
    outliers = df[df[columna] > limite_superior]
    return outliers, limite_superior

# Using the correct column names: 'likesCount' and 'videoViewCount'
outliers_likes, limite_likes = detectar_outliers(df, 'likesCount')
if not outliers_likes.empty:
    print("L√≠mite superior para likes:", limite_likes)
    print("Outliers en likes:")
    display(outliers_likes.head())
else:
    print("No se encontraron outliers para 'likesCount' o la columna no es v√°lida.")

outliers_views, limite_views = detectar_outliers(df, 'videoViewCount') # Corrected from 'views' to 'videoViewCount'
if not outliers_views.empty:
    print("L√≠mite superior para views:", limite_views)
    print("Outliers en views:")
    display(outliers_views.head())
else:
    print("No se encontraron outliers para 'videoViewCount' o la columna no es v√°lida.")
```

Cuando un post tiene valores extremadamente altos en likes o views, usualmente puede estar asociado a contenido patrocinado, colaboraciones con influencers, picos virales o errores de carga en la data. Para determinar si son org√°nicos o pagos, se revisar√≠a: fecha, tipo de contenido, pauta activa, hashtags, presencia de creador invitado, o patrones inusuales en el crecimiento.

Si los incluyes, el promedio y otros indicadores se distorsionan, haciendo que el rendimiento general parezca mejor de lo que realmente es. Si los excluyes, obtienes una visi√≥n m√°s realista y representativa del comportamiento

3. Calidad de la metrica

Un post con muchos likes y 0 comentarios puede indicar que el contenido es altamente visual pero no invita a la conversaci√≥n, que la audiencia es pasiva, que falt√≥ un llamado a la acci√≥n, o incluso que el alcance fue impulsado por pauta sin generar interacci√≥n profunda; tambi√©n podr√≠a levantar sospechas de engagement poco org√°nico si se repite.

1. Definicion de eficiencia

La eficiencia en una estrategia de contenido es qu√© tan bien un post convierte el alcance o la visibilidad en acciones reales: interacci√≥n, conversaci√≥n, clics o cualquier comportamiento que indique que la audiencia no solo vio el contenido, sino que respondi√≥ a √©l.

El engagement rate es mejor indicador que los likes totales porque mide la interacci√≥n relativa al tama√±o de la audiencia. Dos marcas con vol√∫menes distintos no pueden compararse por n√∫meros absolutos, ya que una marca grande siempre tendr√° m√°s likes por simple escala. El engagement rate corrige esa desigualdad y muestra qu√© marca logra que su propia comunidad participe proporcionalmente m√°s.

2. Comparacion entre marcas

```python
import pandas as pd

df["likesCount"] = df["likesCount"].fillna(0)
df["commentsCount"] = df["commentsCount"].fillna(0)
df["videoViewCount"] = df["videoViewCount"].fillna(0)

df["engagement_rate"] = (df["likesCount"] + df["commentsCount"]) / df["videoViewCount"]

df["engagement_rate"] = df["engagement_rate"].replace([float('inf'), -float('inf')], 0)
df["engagement_rate"] = df["engagement_rate"].fillna(0)

resumen = df.groupby("brand")["engagement_rate"].agg(["mean", "median"]).reset_index()
resumen
```

Con un engagement_rate medio muy superior, Shein es m√°s eficiente captando atenci√≥n por publicaci√≥n: logra que una mayor proporci√≥n de su audiencia interact√∫e, incluso si la mediana es 0. Esto indica que algunos posts de Shein generan picos de interacci√≥n muy altos, suficientes para elevar fuertemente su promedio.

con estos datos mirar√≠a con mejores ojos a Shein, porque su capacidad de generar picos de atenci√≥n significa que tiene contenidos capaces de romper el algoritmo y atraer interacci√≥n intensa, algo valioso en crecimiento org√°nico y en construcci√≥n de marca.

3. Distribucion del engagement

El hecho de que la mediana de Shein sea 0 indica que al menos la mitad de sus posts pr√°cticamente no generan interacci√≥n. Pero la media es alta, lo que solo ocurre cuando unos pocos posts tienen engagement muy elevado y jalan todo el promedio hacia arriba.

Temu muestra media y mediana muy cercanas (0.0194 y 0.0186), lo que indica que la mayor√≠a de sus posts rinden parecido, sin grandes picos pero sin fracasos absolutos.

4. Trade-off volumen vs eficiencia

Cuando una marca publica mucho y a√∫n as√≠ su engagement promedio es bajo, es una se√±al de desgaste, la audiencia ve tantas publicaciones que deja de interactuar, o el contenido se vuelve repetitivo y predecible.

Esa estrategia puede ser rentable porque su negocio se basa en frecuencia, rotaci√≥n y descubrimiento constante.

```python
import pandas as pd
import numpy as np

df["likesCount"] = df["likesCount"].fillna(0)
df["commentsCount"] = df["commentsCount"].fillna(0)
df["videoViewCount"] = df["videoViewCount"].fillna(0)

df["engagement_rate"] = (df["likesCount"] + df["commentsCount"]) / df["videoViewCount"]

df["engagement_rate"] = df["engagement_rate"].replace([float('inf'), -float('inf')], 0)
df["engagement_rate"] = df["engagement_rate"].fillna(0)


def categorize_content(caption):
    if isinstance(caption, str):
        caption = caption.lower()
        if 'descuento' in caption or 'oferta' in caption or 'sale' in caption:
            return 'Descuento'
        elif 'haul' in caption or 'compras' in caption or 'coleccion' in caption or 'outfit' in caption:
            return 'Haul/Producto'
        elif 'sostenible' in caption or 'eco' in caption or 'verde' in caption:
            return 'Sostenibilidad'
        elif 'meme' in caption or 'humor' in caption or 'viral' in caption:
            return 'Meme'
        elif 'producto' in caption or 'item' in caption or 'review' in caption or 'unboxing' in caption:
            return 'Producto individual'
        elif 'tip' in caption or 'consejo' in caption or 'guia' in caption:
            return 'Consejo'
        elif 'ugc' in caption or 'usuarios' in caption or 'cliente' in caption:
            return 'UGC'
    return 'Otro'

df['content_type'] = df['caption'].apply(categorize_content)
# ------------------------------------------------------------------

eng_por_tipo = (
    df.groupby(["brand", "content_type"])["engagement_rate"]
      .mean()
      .reset_index()
      .sort_values(["brand", "engagement_rate"], ascending=[True, False])
)

eng_por_tipo
```

Si el contenido con mayor engagement_rate es tambi√©n uno de los que la marca publica con mayor frecuencia, significa que est√° apostando por el formato ganador y optimizando su estrategia.
